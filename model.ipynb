```python
#Dropping critical missing values
walkidx = walkidx.dropna(subset=['fips_code', 'total_population', 'households', 'total_workers', 'total_employment'])
#Excessive missing values
threshold = 0.3 * walkidx.shape[1]
walkidx = walkidx.dropna(thresh=int(threshold))
walkidx
```

```python
#Imputation
#Replace NaN with 0 for acres
area = ['acres_total', 'acres_water', 'acres_land', 'acres_undeveloped']
walkidx[area] = walkidx[area].fillna(0)
#Replace percentages with median
percentage = [col for col in walkidx.columns if 'percent' in col]
for col in percentage:
    walkidx[col] = walkidx[col].fillna(walkidx[col].median())

#Replace employment breakdown with 0 (using total_employment as backup)
employment = [col for col in walkidx.columns if 'employment_' in col]
walkidx[employment] = walkidx[employment].fillna(0)

#Replace urban planning metrics with median
d_metrics = ['res_density_per_acre', 'job_density_per_acre', 'jobs_per_household',
       'land_use_mix', 'land_use_mix_auto', 'arterial_street_density',
       'local_street_density', 'local_to_arterial_ratio',
       'jobs_accessible_walk', 'jobs_accessible_transit', 'pop_near_transit',
       'jobs_near_transit']
for col in d_metrics:
    walkidx[col] = walkidx[col].fillna(walkidx[col].median())
#Geographic data
geo = ['csa_name', 'cbsa_name']
walkidx[geo] = walkidx[geo].fillna('Unincorporated')
walkidx
```

## Data Wrangling

Imputing missing values and adding columns for better analysis

```python
# clean metro area names
walkidx['cbsa_name'] = walkidx['cbsa_name'].str.strip()
        
# Extract primary city (before first comma)
walkidx.insert(1, 'primary_city', walkidx['cbsa_name'].str.split(',').str[0])
        
# Extract state (after last comma)
walkidx.insert(2, 'state', walkidx['cbsa_name'].str.split(',').str[-1].str.strip())
walkidx
```

```python
walkidx = walkidx.drop(columns=['csa_name', 'cbsa_name'])
walkidx
```

```python
#Assign regional categories
northeast = ['CT', 'ME', 'MA', 'NH', 'RI', 'VT', 'NJ', 'NY', 'PA']
midwest = ['IL', 'IN', 'MI', 'OH', 'WI', 'IA', 'KS', 'MN', 'MO', 'NE', 'ND', 'SD']
south = ['DE', 'DC', 'FL', 'GA', 'MD', 'NC', 'SC', 'VA', 'WV', 'AL', 'KY', 'MS', 'TN', 'AR', 'LA', 'OK', 'TX']
west = ['AZ', 'CO', 'ID', 'MT', 'NV', 'NM', 'UT', 'WY', 'AK', 'CA', 'HI', 'OR', 'WA']
def assign_region(state):
    if state in northeast:
        return "Northeast"
    elif state in midwest:
        return "Midwest"
    elif state in south:
        return "South"
    elif state in west:
        return "West"
    else:
        return "Unknown"
walkidx.insert(3, 'region', walkidx['state'].apply(assign_region))
walkidx
```

```python
#Categorize data
walkidx['walkability'] = pd.cut(
        walkidx['national_walkability_index'], 
        bins=[0, 5, 10, 15, 20],
        labels=['Low', 'Medium', 'High', 'Very High'],
        include_lowest=True
    )

walkidx['density'] = pd.qcut(
        walkidx['total_population'] / (walkidx['acres_land'] + 0.1),  # Pop per acre
        q=4,
        labels=['Low Density', 'Medium-Low Density', 
                'Medium-High Density', 'High Density'],
        duplicates='drop'
    )
walkidx['use_diversity'] = pd.cut(
        walkidx['land_use_mix'],
        bins=[0, 0.25, 0.5, 0.75, 1.0],
        labels=['Low Mix', 'Medium Mix', 'High Mix', 'Very High Mix'],
        include_lowest=True
    )
walkidx
```

```python
# boolean flags
walkidx['is_high_walkability'] = walkidx['national_walkability_index'] >= 15
walkidx['is_transit_accessible'] = walkidx['pop_near_transit'] > walkidx['pop_near_transit'].quantile(0.5)
walkidx['is_auto_dependent'] = walkidx['arterial_street_density'] > walkidx['local_street_density']
walkidx['is_metro_area'] = walkidx['cbsa_population'] > 0
walkidx['has_zero_cars'] = walkidx['percent_zero_vehicles'] > 10 
walkidx['is_job_rich'] = walkidx['jobs_per_household'] > 1.5
walkidx['is_mixed_use'] = walkidx['land_use_mix'] > 0.5
walkidx
```

```python
# Classify Block group
def classify_area_type(row):
        if row['national_walkability_index'] >= 15 and row['land_use_mix'] > 0.6:
            return 'Urban Core'
        elif row['national_walkability_index'] >= 10 and row['land_use_mix'] > 0.4:
            return 'Urban Neighborhood'
        elif row['national_walkability_index'] >= 5 and row['cbsa_population'] > 0:
            return 'Suburban'
        elif row['cbsa_population'] > 0:
            return 'Car-Dependent Suburban'
        else:
            return 'Rural/Unincorporated'
    
walkidx['area_type'] = walkidx.apply(classify_area_type, axis=1)
walkidx
```

## Hypothesis Testing

### Question 1
$H_0$ - Arterial street density is independent of walkability  
$H_1$ - Higher arterial street density correlates to lower walkability   
Testing Method: Multiple Linear Regression

```python
import statsmodels.api as sm

y = walkidx['national_walkability_index']
X = walkidx[['arterial_street_density', 'percent_two_plus_vehicles']] 

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

model.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif= pd.DataFrame()

vif["Variable"] = X.columns[1:]
vif["VIF"] = [variance_inflation_factor(X.values, i+1) for i in range(len(X.columns)-1)]

vif
```

```python
sample_idx = np.random.choice(model.resid.index, size=5000, replace=False)
plt.figure(figsize=(10, 6))
plt.scatter(model.fittedvalues[sample_idx], model.resid[sample_idx], alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title(f'Residuals vs Fitted (R² = {model.rsquared:.3f})')
plt.grid(True, alpha=0.3)
plt.show()
```

### Question 3

Testing Method: Multiple Linear Regression
